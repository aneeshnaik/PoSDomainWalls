Here, I'm going to plot the MW plane of satellites, both in an unrotated co-ordinate system and a rotated one.

Imports:

```python
import numpy as np
import matplotlib.pyplot as plt
pi = np.pi
```

```python
def construct_Rz(theta):
    ct = np.cos(theta)
    st = np.sin(theta)
    R = np.array([[ct, -st, 0], [st, ct, 0], [0, 0, 1]])
    return R

def construct_Ry(theta):
    ct = np.cos(theta)
    st = np.sin(theta)
    R = np.array([[ct, 0, st], [0, 1, 0], [-st, 0, ct]])
    return R

def convert_angles(theta, phi, theta0, phi0):
    # convert to Cartesian
    x = np.sin(theta) * np.cos(phi)
    y = np.sin(theta) * np.sin(phi)
    z = np.cos(theta)
    pos = np.stack((x, y, z))
    
    # rotate
    R1 = construct_Rz(-phi0)
    R2 = construct_Ry(-theta0)
    pos_new = np.matmul(R2, np.matmul(R1, pos))
    x_new = pos_new[0]
    y_new = pos_new[1]
    z_new = pos_new[2]

    # get angles
    phi_new = np.arctan2(y_new, x_new)
    theta_new = np.arctan2(np.sqrt(x_new**2 + y_new**2), z_new)
    return theta_new, phi_new
```

Load data

```python
data = np.loadtxt("../sat_coords.txt", skiprows=1, usecols=[1, 2, 3])
l = data[:, 0] * pi / 180
b = data[:, 1] * pi / 180
phi = np.copy(l)
theta = pi / 2 - b
d_kpc = data[:, 2]
```

Convert to Cartesians:

```python
x = d_kpc * np.sin(theta) * np.cos(phi)
y = d_kpc * np.sin(theta) * np.sin(phi)
z = d_kpc * np.cos(theta)
```

```python
plt.scatter(x, z);
```

Rotate into satellite plane

```python
b0 = -10.4 * pi / 180
theta0 = pi / 2 - b0
phi0 = 186.1 * pi / 180

theta_new, phi_new = convert_angles(theta, phi, theta0, phi0)
x_new = d_kpc * np.sin(theta_new) * np.cos(phi_new)
y_new = d_kpc * np.sin(theta_new) * np.sin(phi_new)
z_new = d_kpc * np.cos(theta_new)
```

```python
plt.scatter(x_new, z_new);
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x_new, y_new, z_new);
```
