Here, I want to find out if I can distinguish between on-plane and off-plane satellites by considering a cut in 'vertical energy', somehow defined.

Imports:

```python
import numpy as np
import sys
import matplotlib.pyplot as plt
from os.path import exists
sys.path.append("..")
from src.constants import MSUN, KPC, GYR
from src.simulation import Simulation
from src.satellites import sample_satellites
```

Set up and run sim:

```python
dfile = "20220127_data.npz"
if not exists(dfile):

    # MW and DW params
    M_vir = 1e+12 * MSUN
    c_vir = 10
    a_DW = 5e-11
    l_DW = 10 * KPC

    # sample satellites
    pos, vel = sample_satellites(N_sat=1000, M_vir=M_vir, c_vir=c_vir, f_vesc=0.95)

    # set up simulations
    sim = Simulation(
        pos0=pos, vel0=vel,
        M_vir=M_vir, c_vir=c_vir,
        a_DW=a_DW, l_DW=l_DW
    )

    # run simulation
    sim.run(t_max=3e+18, N_snapshots=5000, dt=1e+12)
    np.savez(dfile,
             pos = sim.positions / KPC, vel = sim.velocities / 1000,
             t = sim.times / GYR)

data = np.load(dfile)
pos = data['pos']
vel = data['vel']
t = data['t']

v = np.linalg.norm(vel, axis=-1)
r = np.linalg.norm(pos, axis=-1)
vz = vel[..., 2]
z = pos[..., 2]
```

Potential:

```python
def calc_phi(z, a_DW, l_DW):
    phi = - 2 * a_DW * l_DW / np.cosh(z / (2 * l_DW))**2
    return phi

U = calc_phi(z * KPC, a_DW=5e-11, l_DW =10*KPC)
```

Kinetic and total energy:

```python
T = 0.5 * (vz * 1000)**2
Ez = T + U
```

```python
zmax = np.max(np.abs(z), axis=0)
plt.scatter(Ez.flatten(), z.flatten(), s=0.5);
```

```python
m1 = np.all(Ez < 0, axis=0)
m2 = Ez[-1] < 0
plt.scatter(pos[-1, :, 0], pos[-1, :, 2]);
plt.scatter(pos[-1, m1, 0], pos[-1, m1, 2]);
plt.xlim(-500, 500)
plt.ylim(-500, 500)
```

```python
plt.scatter(pos[-1, :, 0], pos[-1, :, 2]);
plt.scatter(pos[-1, m2, 0], pos[-1, m2, 2]);
plt.xlim(-500, 500)
plt.ylim(-500, 500)
```
