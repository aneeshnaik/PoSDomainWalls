```python
import matplotlib.pyplot as plt
import numpy as np

# load data
data = np.load("../simulations/sim_noDW.npz")
x0 = data['x']
v0 = data['v']
t0 = data['t']
data = np.load("../simulations/sim_DW.npz")
x1 = data['x']
v1 = data['v']
t1 = data['t']
N_sat = x0.shape[1]
assert x1.shape[1] == N_sat
np.testing.assert_allclose(t0, t1)
```

```python
L = np.cross(x0[-1], v0[-1])
Lx = L[:, 0]
Ly = L[:, 1]
Lz = L[:, 2]
theta = np.arctan2(Lz, np.sqrt(Lx**2 + Ly**2))
phi = np.arctan2(Ly, Lx)

fig = plt.figure()
ax = fig.add_subplot(111, projection='mollweide')
ax.scatter(phi, theta, s=1);
```

```python
L = np.cross(x1[-1], v1[-1])
Lx = L[:, 0]
Ly = L[:, 1]
Lz = L[:, 2]
theta = np.arctan2(Lz, np.sqrt(Lx**2 + Ly**2))
phi = np.arctan2(Ly, Lx)

fig = plt.figure()
ax = fig.add_subplot(111, projection='mollweide')
ax.scatter(phi, theta, s=1);
```

```python
fig = plt.figure(figsize=(15, 15))
ax0 = fig.add_subplot(141, projection='lambert')
ax1 = fig.add_subplot(142, projection='lambert')
ax2 = fig.add_subplot(143, projection='lambert')
ax3 = fig.add_subplot(144, projection='lambert')

L = np.cross(x0[-1], v0[-1])
Lx = L[:, 0]
Ly = L[:, 1]
Lz = L[:, 2]
theta = np.arctan2(Lz, np.sqrt(Lx**2 + Ly**2))
phi = np.arctan2(Ly, Lx)
ax0.scatter(phi, theta, s=0.1)

L = np.cross(x1[-1], v1[-1])
Lx = L[:, 0]
Ly = L[:, 1]
Lz = L[:, 2]
theta = np.arctan2(Lz, np.sqrt(Lx**2 + Ly**2))
phi = np.arctan2(Ly, Lx)
ax1.scatter(phi, theta, s=0.1)

L = np.cross(x0, v0)
Lx = np.average(L[..., 0], axis=0)
Ly = np.average(L[..., 1], axis=0)
Lz = np.average(L[..., 2], axis=0)
theta = np.arctan2(Lz, np.sqrt(Lx**2 + Ly**2))
phi = np.arctan2(Ly, Lx)
ax2.scatter(phi, theta, s=0.1)

L = np.cross(x1, v1)
Lx = np.average(L[..., 0], axis=0)
Ly = np.average(L[..., 1], axis=0)
Lz = np.average(L[..., 2], axis=0)
theta = np.arctan2(Lz, np.sqrt(Lx**2 + Ly**2))
phi = np.arctan2(Ly, Lx)
ax3.scatter(phi, theta, s=0.1)






for ax in fig.axes:
    ax.grid(True)
```
